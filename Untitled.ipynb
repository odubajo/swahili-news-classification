```python

def remove_stopwords(text):
    text = ' '.join(word for word in text.split(' ') if word not in stop_words)
    return text
    
train['content'] = train['content'].apply(remove_stopwords)
```

### tokenization

```python

train['content'] = train['content'].apply(lambda x: word_tokenize(str(x)))
test['content'] = test['content'].apply(lambda x: word_tokenize(str(x)))
```

### lemmatization

```python
lemmatizer = WordNetLemmatizer()

def lemmatization(text):
    text= [lemmatizer.lemmatize(word = x, pos = 'v') for x in text]
    return text

train['content'] = train['content'].apply(lemmatization)
```
